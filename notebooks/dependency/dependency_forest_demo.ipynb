```python
import spacy
from discopy.grammar.dependency import document_to_dependency_diagrams
nlp = spacy.load("en_core_web_sm")
```

```python
tree_doc = nlp("This is my dependency tree!")
trees = document_to_dependency_diagrams(tree_doc) # returns a list
trees[0].draw(figsize=(10,4))
```

```python
forest_doc = nlp("This is my dependency forest! It is made up of lots of dependency trees. Like me! And me!")
forest = document_to_dependency_diagrams(forest_doc)
for tree in forest: tree.draw(figsize=(8,4))
```
