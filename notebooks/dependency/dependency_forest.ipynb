# "Dependency Forest"

```python
import spacy
from discopy.grammar.dependency import dependency_forest
nlp = spacy.load("en_core_web_sm")
```

```python
tree_doc = nlp("This is my dependency tree!")
trees = dependency_forest(tree_doc) # returns a list
trees[0].draw(figsize=(10,4))
```

```python
forest_doc = nlp("This is my dependency forest! It is made up of lots of dependency trees. Like me! And me!")
forest = dependency_forest(forest_doc)
for tree in forest: tree.draw(figsize=(8,4))
```

```python
import functools
functools.reduce(lambda x,y: x@y, forest).draw(figsize=(20,10))
```
